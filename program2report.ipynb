## Approach to Solving Problem
The approach that we took to solving this problem was a divide and conquer strategy which continously narrows down the elements to be sorted by choosing a pivot value and comparing all elements to that pivot value, grouping them by whether they are smaller, equal, or greater than the pivot. At each step, we determine which group can contain the kth smallest element and recursively apply the same process only to that subarray, discarding the remaining elements until the kth smallest element is found and returned.



The illustration above visually shows the process for how we want the selection algorithm to be more efficient. It uses the middle element as the pivot, and then the array is partitioned into three groups: elements less than the pivot, equal to the pivot, and greater than the pivot. By comparing 
k to the sizes of these groups, we can determines which portion will contain the kth smallest element and discards the rest. So in this case, since k is 3 and there are 3 elements smaller than the pivot, we know to just keep that first array. The process is then repeated recursively on the relevant subarray until only one element remains, which will be returned since it is the base case. This demonstrates how the algorithm reduces the problem size at each step instead of sorting the entire array.



The illustration above is demonstrating how the array shrinks at each recursive step until the base case is reached and returned. The structure resembles a binary tree because each pivot creates branches representing the possible partitions. But a decision is made each time the algorithm runs, so only one branch is followed at each step while the others are discarded, illustrating how the divide-and-conquer strategy reduces the problem size efficiently.

## Psuedocode
```text
procedure Selection (A, k)
    //base case
    if(A.length == 1)
        return A[0]
    end if
    
    pivot = middle element of array
    smallerArr = []
    equalArr = []
    greaterArr = []

    for(each x in A)
        if(x == pivot)
            add x to equalArr
        else if(x < pivot)
            add x to smallerArr
        else
            add x to greaterArr
        end if
    end for loop

    if(smallerArr.length >= k)
        return Selection(smallerArr, k)
    else if(smallerArr.length + equalArr.length >= k)
        return pivot
    else
        return Selection(greaterArr, k - (smallerArr.length + equalArr.length))
    end if
end procedure
```
